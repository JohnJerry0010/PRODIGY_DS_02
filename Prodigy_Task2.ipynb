### Problem Statement:



The goal of this project is to perform data cleaning and exploratory data analysis (EDA) on a car dataset to explore the relationships between variables and identify patterns and trends in the data.

##### Step 1: Data Loading

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/hp/OneDrive/Documents/Datasets/car.csv")
```

```python
# Creatimg DataFrame from the loaded data
df = pd.DataFrame(data)

# Displaying the first few rows of the DataFrame
df.head(10)
```

##### Step 2: Data Exploration

```python
# Summary statistics of numerical variables
print(df.describe())
```

```python
# Information about DataFrame including data types and non-null values
print(df.info())
```

##### Step 3: Data Cleaning

```python
df['enginetype'] = df['enginetype'].replace('l', np.nan)
print(df)

```

```python
#to check for duplicate rows
duplicate_rows = df.duplicated()

# Print the duplicate rows
print(duplicate_rows)

```

```python
#to check for the missing values
print(df.isnull().sum())

```

```python
#Removing Uneccessary rows
df = df.dropna()

df = df.drop(columns=['symboling'])
```

```python
#to convert the string values into the int
df['doornumber'] = df['doornumber'].replace({'two': '2','four': '4'})

df['drivewheel'] = df['drivewheel'].replace('4wd', 'fwd')

df['cylindernumber'] = df['cylindernumber'].replace({'two':'2','three':'3','four':'4','five':'5','six':'6','eight':'8','twelve':'12'})
```

##### Step 4: Data Visualization & Interpretation

```python
# Bar Plot: Fuel Type Distribution
fueltype_counts = df['fueltype'].value_counts()

colors = ['lightblue', 'lightgreen']  
plt.bar(fueltype_counts.index, fueltype_counts.values, color=colors)
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Bar Plot: Fuel Type Distribution')
plt.show()
```

The bar plot illustrates the distribution of fuel types, showcasing gas as significantly more common than diesel, with gas vehicles outnumbering diesel vehicles by a large margin.

```python
# Scatter Plot: Highway MPG vs. City MPG
plt.figure(figsize=(8, 6))
plt.scatter(df['citympg'], df['highwaympg'], c=df['enginesize'], cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Price')
plt.title('Highway MPG vs. City MPG')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.show()
```

The scatter plot compares City MPG and Highway MPG, showing a positive correlation where higher City MPG tends to correspond with higher Highway MPG

```python
# Histogram: Car Length Distribution
plt.figure(figsize=(8, 6))
plt.hist(df['carlength'], bins=10, color='purple', edgecolor='black')
plt.title('Car Length Distribution')
plt.xlabel('Car Length')
plt.ylabel('Frequency')
plt.show()
```

The bar graph displays the distribution of car lengths, with the most common length observed around 180 units. Other notable lengths include around 170 and 190 units, but with lower frequencies compared to 180 units.

```python
# Box Plot: Horsepower Distribution for BMW Models
bmw_cars = df[df['CarName'].str.startswith('bmw')]
bmw_models = bmw_cars['CarName'].unique()
box_data = []

for model in bmw_models:
    box_data.append(bmw_cars[bmw_cars['CarName'] == model]['horsepower'].values)

plt.boxplot(box_data, labels=bmw_models)
plt.title('Horsepower Distribution for BMW Models')
plt.ylabel('Horsepower')
plt.grid(True)
plt.show()
```

The box plot illustrates the horsepower distribution across various BMW models. Each model is represented by a box plot, showing the range and distribution of horsepower values. The BMW X3, for instance, exhibits a wide range of horsepower values compared to other models, allowing for comparison between specific horsepower values for each model.

```python
# Pairplot for multiple variables
sns.pairplot(df[['price', 'horsepower', 'enginesize', 'fueltype']], hue='fueltype')
plt.title('Pairplot of Price, Horsepower, Enginesize by Fuel Type')
plt.show()

```

The pair plot visualizes relationships between price, horsepower, and engine size based on fuel type (gas or diesel). It shows that higher horsepower and engine size generally result in higher prices, with diesel vehicles clustering at higher price points compared to gas vehicles with similar attributes.

```python
numerical_cols = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']# Correlation matrix
# Correlation matrix for numerical variables
numerical_df = df[numerical_cols]
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix reveals significant relationships between various car attributes. For instance, there's a strong positive correlation between engine size and horsepower, while highway mpg shows a strong negative correlation with curb weight. Overall, it suggests that certain attributes, like engine size and curb weight, strongly influence others, such as horsepower and fuel efficiency.
